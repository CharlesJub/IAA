## Using the Ameshousing dataset from our in-class examples, run some distributional analysis on Sale_Price, Log(Sale_Price), and Gr_Liv_Area.

### Create histograms of these three variables

```r
gghistogram(ames$Sale_Price, y="density", fill = "blue", add_density=TRUE)
```

```r
gghistogram(ames$Gr_Liv_Area, y="density", fill = "red", add_density=TRUE)
```
